```python
# MINI SEARCHING TOOL PROJECT

print("""Which tool you want to use from below?
         1.COPY PASTE TOOL
         2.STACK OVERFLOW SEARCH TOOL
         3.WIKIPEDIA SEARCH TOOL
         4.QR CODE GENERATOR TOOL
         5.URL EXTRACTION TOOL
         6.URL SHORTENER TOOL
         7.GOOGLE SEARCH TOOL
         8.DICTIONARY TOOL
       """   )

index_num = int(input("Enter The number: "))

if index_num == 1:
    import pyperclip
    element1 = input("Enter what u want to copy - ")
    print("paste using ctrl+V below")
    pyperclip.copy(element1)
    pyperclip.paste()


elif index_num == 2:
    import howdoi
    print("Enter howdoi along with what you want to search in stack")


elif index_num == 3:
    import wikipedia
    element2 = input("Enter what do you want to search in wikipedia - ")
    query=wikipedia.page(element2)
    print(query.summary)

elif index_num == 4:
    import pyqrcode
    import png
    def qrcode():
        print("Give the input to generate")
        q = pyqrcode.create(input())
        q.png('QRcode.png',scale=10)
        print("QR code generated")
    if __name__=='__main__':
        qrcode()

elif index_num == 5:
    from urllib.request import urlopen
    url = input("Enter the URL - ")
    page = urlopen(url)
    print("""What do you want?
             a.Header
             b.sourcecode""")
    option = input("Enter your option a or b - ")
    if option == "a":
        print(page.headers) 
    else:
        sourcecode = page.read()
        print(sourcecode)

elif index_num == 6:
    import pyshorteners
    url = input("Enter The URL: ")
    shortener = pyshorteners.Shortener()
    a = shortener.tinyurl.short(url)
    print(a)


elif index_num == 7:
    from googlesearch import search
    query = input("Enter what you want to search on google - ")
    for i in search(query,start=0,pause=2):
        print(i)


elif index_num == 8:
    from PyDictionary import PyDictionary
    dictionary = PyDictionary()
    element3 = input("Enter a word - ")
    print(dictionary.meaning(element3))

else:
    print("INVALID INDEX")
```
